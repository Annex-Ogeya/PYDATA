# Common Data Visualizations in Python

This notebook demonstrates **10 common data visualizations** using `matplotlib`, `seaborn`, and `pandas`.

Visualizations included:
1. Bar Chart
2. Histogram
3. Pie Chart
4. Line Chart
5. Scatter Plot
6. Box Plot
7. Heatmap
8. Stacked Area Chart
9. Bubble Chart
10. Violin Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Generate sample data
np.random.seed(42)
categories = ["A", "B", "C", "D"]
values = [23, 45, 12, 36]
time = pd.date_range("2021-01", periods=12, freq="M")
trend = np.cumsum(np.random.randn(12)) + 50
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(50, 500, size=50)
groups = np.random.choice(["G1", "G2", "G3"], size=50)

# Apply style
plt.style.use("seaborn-v0_8")
```

## Bar Chart

```python
plt.figure(figsize=(6,4))
plt.bar(categories, values, color="skyblue")
plt.title("Bar Chart: Category Comparison")
plt.show()
```

## Histogram

```python
plt.figure(figsize=(6,4))
plt.hist(np.random.randn(200), bins=15, color="lightgreen", edgecolor="black")
plt.title("Histogram: Data Distribution")
plt.show()
```

## Pie Chart

```python
plt.figure(figsize=(6,6))
plt.pie(values, labels=categories, autopct="%1.1f%%", colors=["#ff9999","#66b3ff","#99ff99","#ffcc99"])
plt.title("Pie Chart: Proportion of Categories")
plt.show()
```

## Line Chart

```python
plt.figure(figsize=(6,4))
plt.plot(time, trend, marker="o", color="purple")
plt.title("Line Chart: Trend Over Time")
plt.xticks(rotation=45)
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(6,4))
plt.scatter(x, y, alpha=0.7, color="orange")
plt.title("Scatter Plot: Relationship Between Two Variables")
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=[np.random.randn(100), np.random.randn(100)+2, np.random.randn(100)+5])
plt.title("Box Plot: Distribution and Outliers")
plt.show()
```

## Heatmap

```python
plt.figure(figsize=(6,4))
data = np.random.rand(6,6)
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("Heatmap: Value Intensity Across Dimensions")
plt.show()
```

## Stacked Area Chart

```python
plt.figure(figsize=(6,4))
data = np.random.rand(12,3)
plt.stackplot(time, data[:,0], data[:,1], data[:,2], labels=["X","Y","Z"], alpha=0.8)
plt.legend()
plt.title("Stacked Area Chart: Composition Over Time")
plt.xticks(rotation=45)
plt.show()
```

## Bubble Chart

```python
plt.figure(figsize=(6,4))
plt.scatter(x, y, s=sizes, alpha=0.5, c=sizes, cmap="viridis")
plt.title("Bubble Chart: Three Variables")
plt.show()
```

## Violin Plot

```python
plt.figure(figsize=(6,4))
sns.violinplot(x=groups, y=np.random.randn(50), palette="Set2")
plt.title("Violin Plot: Distribution Across Groups")
plt.show()
```
